```python
ax = sns.countplot(x = 'Gender',data = df )

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
ax = sns.countplot(data = df, x = "Age Group", hue = "Gender") 

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Assuming df and sales_age are defined correctly
data = {
    "Age Group": ["18-25", "26-35", "36-45", "46-55", "56+"],
    "Amount": [1000, 1500, 1100, 1300, 900]
}

sales_age = pd.DataFrame(data)

# Define a color palette
colors = sns.color_palette("husl", n_colors=len(sales_age))  

# Plotting
sns.set(rc={'figure.figsize':(10,5)})  
sns.barplot(x="Age Group", y='Amount', data=sales_age, palette=colors)
plt.xlabel("Age Group", fontsize=14)
plt.ylabel("Total Amount", fontsize=14)
plt.title("Total Amount vs Age Group", fontsize=16)
plt.show()
```

From above graphs we can see that most of the buyers are of age group between 26-35yrs female

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df and sales_state are defined as in your code

# Group by state and sum the number of orders
sales_state = df.groupby(["State"], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

# Define a color palette
colors = sns.color_palette("tab10")  # Using the 'tab10' palette for 10 different colors

# Plotting
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state, x="State", y='Orders', palette=colors)
plt.show()
```

From ab graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively

```python
# Assuming df is your DataFrame

# Set figure size
sns.set(rc={'figure.figsize':(20,5)})

# Define a color palette
palette = sns.color_palette("tab10")  # Using the 'tab10' palette for 10 different colors

# Create countplot with specified palette
ax = sns.countplot(data=df, x='Occupation', palette=palette)

# Add labels to the bars
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha='center', va='center', 
                   xytext=(0, 10), 
                   textcoords='offset points')

# Show the plot
plt.show()
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
                                                                                     
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = "Occupation",y= 'Amount')                                                                                     
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```
